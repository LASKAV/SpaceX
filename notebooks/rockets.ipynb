```python
import requests
import json
import pandas as pd

URL = "https://api.spacexdata.com/v4/rockets"

response = requests.get(URL).json()
rockets_data = json.dumps(response, indent=4)
print(rockets_data)
```

```python
response = requests.get(URL)
if response.status_code == 200:
    data = response.json()
    parsed_data = [
        {
            "id": item["id"],
            "name": item["name"],
            "type": item["type"],
            "active": item["active"],
            "country": item["country"],
            "engine_numbers": item["engines"]["number"],
            "engine_type": item["engines"]["type"],
            "landing_legs": item["landing_legs"]["number"]
        }
        for item in data
    ]
    df_rockets = pd.DataFrame(parsed_data)

df_rockets

```
