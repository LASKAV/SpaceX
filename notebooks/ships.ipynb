```python
import requests
import json
import pandas as pd

URL = "https://api.spacexdata.com/v4/ships"

response = requests.get(URL).json()
ships_data = json.dumps(response, indent=4)
print(ships_data)
```

```python
response = requests.get(URL)
if response.status_code == 200:
    data = response.json()
    parsed_data = [
        {
            "id": item["id"],
            "legacy_id": item["legacy_id"],
            "type": item["type"],
            "roles": item["roles"],
            "imo": item["imo"],
            "mmsi": item["mmsi"],
            "abs": item["abs"],
            "class": item["class"],
            "mass_kg": item["mass_kg"],
            "mass_lbs": item["mass_lbs"],
            "year_built": item["year_built"],
            "home_port": item["home_port"],
        }
        for item in data
    ]
    df_ships = pd.DataFrame(parsed_data)

df_ships
```
